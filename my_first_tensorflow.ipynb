```python
import tensorflow as tf
import numpy as np
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)), #expected share of neuron
    tf.keras.layers.Dense(units=1) #only one neuron, Dense
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error') #2 functions to know
#sgd = stochastic gradient descent
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=500) #train the model using the function "fit" it will go through the loop 500 times.
```

```python
print(f"model.predicted: {model.predict(np.array([10.0]), verbose=0).item():.5f}")
```
